```python
import pyvista as pv
```

```python
# Load the results
results = pv.read("results.xdmf")
```

```python
# Plot the velocity field
plotter = pv.Plotter()
plotter.add_mesh(results, scalars="U", show_edges=True, cmap="viridis")
plotter.add_title("Velocity Field")
plotter.show()
```

```python
# Plot the pressure field
plotter = pv.Plotter()
plotter.add_mesh(results, scalars="P", show_edges=True, cmap="coolwarm")
plotter.add_title("Pressure Field")
plotter.show()
```

```python
# Plot the temperature field
plotter = pv.Plotter()
plotter.add_mesh(results, scalars="T", show_edges=True, cmap="plasma")
plotter.add_title("Temperature Field")
plotter.show()
```

```python
# Create a plotter with subplots
plotter = pv.Plotter(shape=(1, 3))

# Velocity field
plotter.subplot(0, 0)
plotter.add_mesh(results, scalars="U", show_edges=True, cmap="viridis")
plotter.add_title("Velocity Field")

# Pressure field
plotter.subplot(0, 1)
plotter.add_mesh(results, scalars="P", show_edges=True, cmap="coolwarm")
plotter.add_title("Pressure Field")

# Temperature field
plotter.subplot(0, 2)
plotter.add_mesh(results, scalars="T", show_edges=True, cmap="plasma")
plotter.add_title("Temperature Field")

# Show the plotter
plotter.show()
```
